**Section 11 — Visualization Tasks
Create charts for:
Hiring distribution (pie chart)
CGPA distribution (histogram)
Skills score vs hiring (boxplot)
Internships vs hiring (bar chart)
Education level vs hiring (bar chart)**

```python
hired_counts = df['hired'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(hired_counts, labels=['Not Hired', 'Hired'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title("Proportion of Hired vs Not Hired Candidates")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['cgpa'], bins=10, kde=True, color='skyblue')
plt.xlabel("CGPA")
plt.ylabel("Number of Candidates")
plt.title("CGPA Distribution of Candidates")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='hired', y='skills_score', data=df, palette=['#ff9999','#66b3ff'])
plt.xlabel("Hired (0 = No, 1 = Yes)")
plt.ylabel("Skills Score")
plt.title("Skills Score: Hired vs Not Hired Candidates")
plt.show()
```

```python
avg_internships = df.groupby('hired')['internships'].mean()

# Plot
plt.figure(figsize=(7,5))
sns.barplot(x=avg_internships.index, y=avg_internships.values, palette=['#ff9999','#66b3ff'])
plt.xlabel("Hired (0 = No, 1 = Yes)")
plt.ylabel("Average Number of Internships")
plt.title("Average Internships: Hired vs Not Hired Candidates")

plt.show()
```

```python
hiring_rate_edu_level = df.groupby('education_level')['hired'].mean() * 100

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=hiring_rate_edu_level.index, y=hiring_rate_edu_level.values, palette="viridis")
plt.xlabel("Education Level")
plt.ylabel("Hiring Rate (%)")
plt.title("Hiring Rate by Education Level")
plt.ylim(0, 100)  # optional for consistent percentage scale
plt.show()
```

**1️⃣ Does CGPA strongly affect hiring?
Analysis of the dataset shows that candidates with higher CGPA generally have a greater chance of being hired. The average CGPA of hired candidates is noticeably higher than that of non-hired candidates. However, CGPA alone is not a definitive factor—some candidates with lower CGPA are still hired, which indicates that while academic performance matters, companies also consider practical skills, experience, and other attributes when making hiring decisions.**

**2️⃣ Do internships increase hiring chances?
Internships appear to have a strong impact on hiring probability. Candidates who have completed more internships are more likely to be hired, as reflected by the higher average number of internships among hired candidates. This suggests that hands-on exposure to real-world projects and industry experience significantly boosts a candidate’s attractiveness to employers.**

**3️⃣ Do candidates with higher skills scores get hired more?
Yes. Candidates with higher technical and soft skills scores are disproportionately more likely to be hired. The distribution of skills scores shows that top-performing candidates in both areas have a much higher hiring probability. This emphasizes that practical skills and candidate competency are crucial factors in hiring decisions.**

**4️⃣ Does resume length matter?
Resume length has some correlation with hiring, but it is more nuanced. Candidates with resumes longer than the average plus one standard deviation are more likely to be hired, indicating that a detailed resume reflecting skills, projects, and experience can positively influence hiring, but excessively long resumes do not necessarily guarantee selection.**

**5️⃣ What combination of factors leads to hiring?
The strongest predictor of hiring is candidate strength, which combines skills score, soft skills, projects, and internships. Hired candidates consistently have higher overall candidate strength than non-hired ones. Experience, number of projects, internships, and CGPA all contribute, but it is the combination of hands-on experience, technical skills, and soft skills that ultimately drives hiring decisions. Education level has some influence, but practical ability and demonstrated competence dominate the decision-making process.**

**✅ Summary Insight:
Hiring is multifactorial. While CGPA and education are relevant, practical experience, skills, projects, internships, and overall candidate strength are the key drivers. Candidates who focus on developing both technical and soft skills, gaining industry experience, and building strong resumes have the highest chances of being hired.**
